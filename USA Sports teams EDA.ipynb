```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nhl_df=pd.read_csv("nhl.csv")
cities=pd.read_html("wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def nhl_correlation(): 
    cities['NHL']=cities['NHL'].str.extract('([a-zA-Z0-9 \.]*)')
    nhl_df['team']=nhl_df['team'].str.extract('([a-zA-Z0-9 \.]*)')
    nhl_df2=nhl_df[nhl_df['year']==2018]
    cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
    cities['NHL'][cities['NHL']=='RangersIslandersDevils']='Rangers'
    cities['NHL'][cities['NHL']=='Rangers']='Rangers Islanders Devils'
    cities['NHL'][cities['NHL']=='KingsDucks']='Kings'
    cities['NHL'][cities['NHL']=='Kings']='Kings Ducks'
    cities['NHL']=cities['NHL'].str.extract('([a-zA-Z\s]+)')
    cities.dropna(subset=['NHL','Population'],inplace=True)
    nhl_df2=nhl_df2[1:]
    nhl_df2['W']=nhl_df2['W'].str.extract('([0-9]+)')
    nhl_df2['L']=nhl_df2['L'].str.extract('([0-9]+)')
    nhl_df2['L']=nhl_df2['L'].astype(float)
    nhl_df2.dropna(subset=['L','W'],inplace=True)
    cities['ratio']=0
    for y in cities['NHL']:
        Win=0
        Loss=0
        for r in y.split():
            n=cities.loc[cities['NHL']==y].index.values
            for x in nhl_df2['team']:
                 m=nhl_df2.loc[nhl_df2['team']==x].index.values
                 if r in x and len(y)>0:
            #print("y:{}, x:{}".format(y,x))
                    Win=Win+float(nhl_df2.loc[m,'W'].values[0])
                    Loss=Loss+float(nhl_df2.loc[m,'L'].values[0])
                    ratio=Win/(Win+Loss)
                    cities.loc[n,'ratio']=ratio
                    break
    cities['Population']=cities['Population'].astype(float)
    
    population_by_region = list(cities['Population']) # pass in metropolitan area population from cities
    win_loss_by_region = list(cities['ratio']) # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return cities['Population'].corr(cities['ratio'])
nhl_correlation()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("mlb.csv")
cities=pd.read_html("wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]

def mlb_correlation(): 
    mlb_df2=mlb_df[mlb_df['year']==2018]
    cities.columns = cities.columns.str.strip()
    assert('Population (2016 est.)[8]' in cities.columns)
    cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
    cities['MLB']=cities['MLB'].str.replace('\[.*\]','',regex=True)
    cities['MLB']=cities['MLB'].str.extract('([A-Za-z0-9\s]+)')
    cities.dropna(subset=['MLB'],inplace=True)
    cities['MLB'][cities['MLB']=='YankeesMets']='Yankees,Mets'
    cities['MLB'][cities['MLB']=='CubsWhite Sox']='Cubs,White Sox'
    cities['MLB'][cities['MLB']=='DodgersAngels']='Dodgers,Angels'
    cities['MLB'][cities['MLB']=='GiantsAthletics']='Giants,Athletics'
    cities['MLB_ratio']=0
    for y in cities['MLB']:
        Win=0
        Loss=0
        for r in y.split(','):
            n=cities.loc[cities['MLB']==y].index.values
            for x in mlb_df['team']:
                    m=mlb_df.loc[mlb_df['team']==x].index.values
                    if r in x and len(y)>0:
                #print("y:{}, x:{}".format(y,x))
                        Win=Win+float(mlb_df.loc[m,'W'].values[0])
                        Loss=Loss+float(mlb_df.loc[m,'L'].values[0])
                        ratio=Win/(Win+Loss)
                        cities.loc[n,'MLB_ratio']=ratio
                        break
    cities['Population']=cities['Population'].astype(int)
    population_by_region = list(cities['Population']) # pass in metropolitan area population from cities
    win_loss_by_region = list(cities['MLB_ratio']) # pass in win/loss ratio from mlb_df in the same order as cities["Metropolitan area"]
    # resultDf = pd.DataFrame(cities[["Metropolitan area", "Population", "MLB_ratio"]], copy=True)
    # print(resultDf.sort_values(by='Metropolitan area'))
    assert len(population_by_region) == len(win_loss_by_region), "Q3: Your lists must be the same length"
    assert len(population_by_region) == 26, "Q3: There should be 26 teams being analysed for MLB"

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
mlb_correlation()
```
